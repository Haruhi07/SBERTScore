```python
def compute_metric(list1, list2):
    #list1 is target
    TP = 0
    TN = 0
    FP = 0
    FN = 0
    for l1, l2 in zip(list1, list2):
        if l1 == 1:
            if l2 == 1:
                TP = TP + 1
            elif l2 == 0:
                FN = FN + 1
        if l1 == 0:
            if l2 == 1:
                FP = FP + 1
            elif l2 == 0:
                TN = TN + 1
    return TP/len(list1), TN/len(list1), FP/len(list1), FN/len(list1)
```
